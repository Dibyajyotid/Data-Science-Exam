# instagram reach analysis

```python
#importing libraries and dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv("Instagram_data.csv", encoding = 'latin1')
df.head()
```

### Removing Null values if any

```python
df.isnull().sum()
```

### Analysing Reach from Home and showing in a plot graph

```python
plt.figure(figsize=(10, 8))
plt.title("Distribution of Impressions From Home")
sns.distplot(df['From Home'])
plt.show()
```

### Analysing Reach from Hashtags and showing in a plot graph

```python
plt.figure(figsize=(10, 8))
plt.title("Distribution of Impressions From Hashtags")
sns.distplot(df['From Hashtags'])
plt.show()
```

### Analysing Reach from Explore and showing in a plot graph

```python
plt.figure(figsize=(10, 8))
plt.title("Distribution of Impressions From Explore")
sns.distplot(df['From Explore'])
plt.show()
```

### Analysing Reach from Others and showing in a plot graph

```python
plt.figure(figsize=(10, 8))
plt.title("Distribution of Impressions From Others")
sns.distplot(df['From Other'])
plt.show()
```

### Analysing Impressions on Instagram

```python
home = df["From Home"].sum()
hashtags = df["From Hashtags"].sum()
explore = df["From Explore"].sum()
other = df['From Other'].sum()

labels = ['From Home','From Hashtags','From Explore','Other']
values = [home, hashtags, explore, other]

fig = px.pie(df, values=values, names=labels, 
             title='Impressions on Instagram Posts', hole=0.5)
fig.show()
```

### Analysing the number of followers gained from profile visits from a post

```python
conversion_rate = (df["Follows"].sum() / df["Profile Visits"].sum()) * 100
print(conversion_rate)
```
